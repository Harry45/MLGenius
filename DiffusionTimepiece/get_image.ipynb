```python
from utils.helpers import dill_save, dill_load 
import os
from urllib.parse import urlparse
import pandas as pd 
```

```python
cols = ['img_name', 'description']
dataframe = pd.DataFrame(columns = cols)
img_folder = 'images/'
```

```python
%%capture
for i in range(889):
    watchname = f'watch_{i}.pkl'
    if os.path.isfile('data/' + watchname):
        watch = dill_load('data/watchinfo', f'watch_{i}')
        img_url = watch['image']
        if img_url is not None:
            a = urlparse(img_url)
            img_name = os.path.basename(a.path)
            filename = img_folder + '/' + img_name # get the full path of the file
            if os.path.exists(filename):
                os.remove(filename)
            os.system(f'wget {img_url} -P {img_folder}')
            df = pd.DataFrame([img_name, watch['description']]).T
            df.columns = cols
            dataframe = pd.concat([dataframe, df])
```

```python
dataframe.reset_index(drop=True, inplace=True)
```
